```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data
df = pd.read_csv('Kelas D_Happiness and Satisfaction.csv', sep=';', decimal=',')

# Menghitung standar deviasi tanpa pembulatan
agg_df = df.groupby('Region')['Happy Score (0-10)'].std().reset_index()

# Membuat Horizontal Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(data=agg_df, x='Happy Score (0-10)', y='Region', color='skyblue')
plt.title('Kategori A: Standar Deviasi Happy Score per Region')
plt.xlabel('Standar Deviasi Happy Score (0-10)')
plt.ylabel('Region')

plt.tight_layout()
# Menyimpan gambar dengan resolusi tinggi (hapus tanda pagar untuk mengaktifkan)
# plt.savefig('Grafik_A_Agregasi.png', dpi=300)
plt.show()
```
