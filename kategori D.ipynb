```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data
df = pd.read_csv('Kelas D_Happiness and Satisfaction.csv', sep=';', decimal=',')

# Membuat Histogram dengan garis KDE (Kernel Density Estimate)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='median_income (&)', kde=True, color='purple')
plt.title('Kategori D: Distribusi Median Pendapatan Dunia')
plt.xlabel('Median Income')
plt.ylabel('Frekuensi')

plt.tight_layout()
# plt.savefig('Grafik_D_Distribusi.png', dpi=300)
plt.show()
```
