```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data
df = pd.read_csv('Kelas D_Happiness and Satisfaction.csv', sep=';', decimal=',')

# Memfilter region 'Asia' atau 'Europe' dan skor > 60
region_filter = df['Region'].str.contains('Asia|Europe', case=False, na=False)
satisfaction_filter = df['adjusted_satisfaction (0-100)'] > 60
filter_df = df[region_filter & satisfaction_filter]

# Membuat Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(data=filter_df, x='Country', y='adjusted_satisfaction (0-100)', color='coral')
plt.title('Kategori B: Negara Asia/Eropa dengan Kepuasan > 60')
plt.xlabel('Negara')
plt.ylabel('Adjusted Satisfaction (0-100)')
plt.xticks(rotation=45, ha='right') # Memutar nama negara agar tidak bertumpuk

plt.tight_layout()
# plt.savefig('Grafik_B_Filter.png', dpi=300)
plt.show()
```
